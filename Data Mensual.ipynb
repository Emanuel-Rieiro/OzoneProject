```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt

# Open the file
file_path = "MERRA2_400.tavgM_2d_slv_Nx.202410.nc4"
nc_data = Dataset(file_path, mode='r')

# Extract data
variable = nc_data.variables['Var_U250']  # Replace with actual variable name
lat = nc_data.variables['lat'][:]
lon = nc_data.variables['lon'][:]

# Reduce to 2D (e.g., first time step, first level)
variable_2d = variable[0, :, :]  # Select first time and level

# Plot the data
plt.figure(figsize=(10, 6))
plt.contourf(lon, lat, variable_2d, cmap='viridis')
plt.colorbar(label='Variable Units')
plt.title('Visualization of Variable')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Close the file
nc_data.close()
```
